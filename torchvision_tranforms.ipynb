```python
import torch
import torchvision
import matplotlib.pyplot as plt

# load out image into memory
image_path = "/workspace/lecture_code_blocks/assets/sample_image.jpg"
m_raw_img = torchvision.io.read_image(image_path,)
```

```python
# matplotlib expects data (H,W,C), torchvision outputs (C,H,W)
plt.imshow(m_raw_img.permute(1, 2, 0))
```

```python
angle = 180
m_img = torchvision.transforms.RandomRotation(angle)(m_raw_img)
plt.imshow(m_img.permute(1, 2, 0))
```

```python
tranformed_image_size = 256
m_img = torchvision.transforms.CenterCrop(tranformed_image_size)(m_raw_img)
plt.imshow(m_img.permute(1, 2, 0))
```

```python
random_crop_size = 512
random_crop = torchvision.transforms.RandomCrop(random_crop_size)
m_img = random_crop(m_raw_img)
plt.imshow(m_img.permute(1, 2, 0))
```

Lets Chain together a set of transform

```python
transforms = torch.nn.Sequential(
    torchvision.transforms.CenterCrop(tranformed_image_size),
    torchvision.transforms.RandomRotation(angle),
    torchvision.transforms.RandomGrayscale(p=0.1),
)

m_img = transforms(m_raw_img)
plt.imshow(m_img.permute(1, 2, 0))
```
